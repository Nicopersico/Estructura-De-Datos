```python
import networkx as nx
import matplotlib.pyplot as plt

G_ejemplo = nx.Graph()
G_ejemplo.add_edges_from([("A", "B"), ("A", "D"), ("B", "D"), ("B", "E"), ("D", "C"), ("E", "D")])

# Mostramos el recorrido BFS desde 'A'
orden_bfs = bfs(G_ejemplo, "A")
print("Recorrido BFS:", orden_bfs)

# Graficamos el grafo
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G_ejemplo)
nx.draw(G_ejemplo, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=400)
plt.show()
```
